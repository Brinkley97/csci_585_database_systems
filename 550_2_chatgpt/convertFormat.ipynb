```python
LatexContent = '''\\documentclass{scrartcl}
                        \\usepackage{graphicx}
                        \\begin{document}
                            {\\sffamily %(Fn)s \\textsc{%(Ln)s}}
                                \\newline
                            \\section{Phone}
                            {\\tiny Phone number: %(Ph)s}
                                \\newpage
                            %(Ot)s
                   \\end{document}'''

###== Look at the database ==##
# open the database into python
my_db_file = open("../datasets/project_2/sampleData.csv","r")

# read the database
my_db = csv.reader(my_db_file, delimiter=',',skipinitialspace=True)

###== TeX files processing and generating ==###
#skip the header of the database
# my_db.next() 
next(my_db)

#then for each row of the database
for row in my_db :
        ## Assign the items of the row to the variables that will fill up the 
        ##    blanks of the LaTeX code
        ID = str(row[0])            #caution, first item of a row = index '0'
        LastName = str(row[1])
        FirstName = str(row[2])
        Phone = str(row[3])
        Other = str(row[4])

            #define the TeX file name
        TexFileName = ID + '.tex'

        ## create a new LaTeX file with the blanks filled
            #create a new file
            
            # save location
        save_file  = '../datasets/project_2/save_texs/' + TexFileName
        TexFile = open(save_file,'w')

            #fill the blanks with the previously read informations
        TexFile.write(LatexContent %{"Id" : ID, "Fn" : FirstName, 
        "Ln" : LastName, "Ph" : Phone, "Ot" : Other })

            #close the file
        TexFile.close()
        
        print("Process New File")

        subprocess.run(["pdflatex", save_file])
        
    
        print()
        
#close the database file
my_db_file.close()
```

```python
save_pdfs = '../datasets/project_2/save_pdfs/'
save_auxs = '../datasets/project_2/save_auxs/'
save_logs = '../datasets/project_2/save_logs/'
```

```python
def move_files(save_pdfs, save_auxs, save_logs):
   
    dir_list = os.listdir(os.curdir)

    for file in dir_list:
        print("File to move", file)
        file_name = os.path.splitext(file)[0]
        file_extension = os.path.splitext(file)[1]
        # print(file_extension)
        
        if file_extension == ".pdf":
            shutil.move(file, save_pdfs)
        if file_extension == ".aux":
            shutil.move(file, save_auxs)
        if file_extension == ".log":
            shutil.move(file, save_logs)
        print()
```

```python
move_files(save_pdfs, save_auxs, save_logs)
```
