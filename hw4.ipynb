```python
import numpy as np
import pandas as pd
```

```python
R=6
r=1
a=8
t = 0
curve = 12*np.pi
curve

# tt_coords = (,)
```

```python
x_coords = []
y_coords = []
while t < curve:
    # print(t)
    x = (R+r)*np.cos((r/R)*t) - a*np.cos((1+r/R)*t) - 118.28538
    x_coords.append(x)
    y = (R+r)*np.sin((r/R)*t) - a*np.sin((1+r/R)*t) + 34.0205
    y_coords.append(y)
    t = t + 0.01
```

```python
x_long_coords = np.array([x_coords]).T
# x_long_coords

x_df = pd.DataFrame(x_long_coords, columns=['longtitude'])
# x_df
```

```python
y_lat_coords = np.array([y_coords]).T
# y_lat_coords

y_df = pd.DataFrame(y_lat_coords, columns=['latitude'])
# y_df
```

```python
x_y_df = pd.concat([x_df, y_df], axis=1)
x_y_df
```
